```python
import pandas as pd

# 원본 데이터
data = {
    '연도': [2020, 2020, 2019, 2019, 2018, 2018, 2017, 2017, 2016, 2016, 2015, 2015, 2014, 2014],
    '구분': ['발생건수', '검거건수', '발생건수', '검거건수', '발생건수', '검거건수', '발생건수', '검거건수', '발생건수', '검거건수', '발생건수', '검거건수', '발생건수', '검거건수'],
    '해킹(계정도용)': [1067, 117, 751, 118, 721, 160, 1160, 619, 758, 161, 1062, 185, 567, 130],
    '해킹(단순침입)': [1621, 217, 1458, 277, 1003, 241, 904, 219, 714, 194, 809, 191, 760, 226],
    '해킹(자료유출)': [130, 46, 114, 41, 114, 41, 67, 36, 71, 38, 69, 36, 116, 81],
    '해킹(자료훼손)': [358, 168, 341, 120, 340, 142, 299, 116, 304, 144, 307, 112, 205, 103],
    '서비스거부공격': [25, 10, 35, 14, 20, 14, 43, 28, 192, 164, 40, 19, 26, 16]
}

# 데이터프레임 생성
Cybercrime_data = pd.DataFrame(data)

# 발생건수와 검거건수를 분리
occur_data = Cybercrime_data[Cybercrime_data['구분'] == '발생건수'].reset_index(drop=True)
arrest_data = Cybercrime_data[Cybercrime_data['구분'] == '검거건수'].reset_index(drop=True)

# 검거율 계산
arrest_rate = (arrest_data.iloc[:, 2:] / occur_data.iloc[:, 2:]) * 100

# 새로운 데이터프레임 생성
new_data = {
    '연도': occur_data['연도'],
    '해킹(계정도용)_발생건수': occur_data['해킹(계정도용)'],
    '해킹(단순침입)_발생건수': occur_data['해킹(단순침입)'],
    '해킹(자료유출)_발생건수': occur_data['해킹(자료유출)'],
    '해킹(자료훼손)_발생건수': occur_data['해킹(자료훼손)'],
    '서비스거부공격_발생건수': occur_data['서비스거부공격'],
    '해킹(계정도용)_검거건수': arrest_data['해킹(계정도용)'],
    '해킹(단순침입)_검거건수': arrest_data['해킹(단순침입)'],
    '해킹(자료유출)_검거건수': arrest_data['해킹(자료유출)'],
    '해킹(자료훼손)_검거건수': arrest_data['해킹(자료훼손)'],
    '서비스거부공격_검거건수': arrest_data['서비스거부공격'],
    '해킹(계정도용)_검거율(%)': arrest_rate['해킹(계정도용)'],
    '해킹(단순침입)_검거율(%)': arrest_rate['해킹(단순침입)'],
    '해킹(자료유출)_검거율(%)': arrest_rate['해킹(자료유출)'],
    '해킹(자료훼손)_검거율(%)': arrest_rate['해킹(자료훼손)'],
    '서비스거부공격_검거율(%)': arrest_rate['서비스거부공격']
}

new_dataframe = pd.DataFrame(new_data)

display(new_dataframe)
```
