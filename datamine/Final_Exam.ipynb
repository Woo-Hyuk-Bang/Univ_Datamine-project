```python
"""
2.	데이터 준비,백업 및 보기
"""

# 관련 라이브러리 준비
import pandas as pd                       # pandas
import matplotlib as mpl                  # 맷플롯립
import matplotlib.pylab as plt            # 맷플롯립
import seaborn as sns                     # 시본
import numpy as np                        # 넘파이
import matplotlib.font_manager as fm      # 한글폰트 로드

plt.rc('font', family='Malgun Gothic')    # 맑음고딕 한글폰트 준비

# 준비한 csv의 경로를 file_path에 저장
file_path= 'C:/Users/User/Desktop/Project/datamine/Police_crime_2015.csv'

df = pd.read_csv(file_path, encoding='euc-kr') # csv 읽기 및 데이터프레임 변환, 한글이 포함되었기 때문에 인코딩 한국어 포함

# 데이터 원본 백업하기
Police_crime_copy_backup = df.copy()

# to_csv 메소드로 내장 데이터프레임을 csv 파일로 저장
# 경로는 C:/Users/User/Desktop/Project/datamine
df.to_csv('C:/Users/User/Desktop/Project/datamine/Police_crime_2015_backup.csv')

# 저장 파일 확인하기
file_path= 'C:/Users/User/Desktop/Project/datamine/Police_crime_2015_backup.csv'
Police_crime_2015_backup = pd.read_csv(file_path)
Police_crime_2015_backup  #데이터 로드하기

df.info()          # 데이터 프레임의 기본 정보 출력
df.describe()      # 데이터프레임의 기초 통계 확인
df.dtypes          # 데이터 유형 확인
df.loc[0]          # 데이터 행 인덱스 확인
df.head()          # 상위 데이터 추출
df.tail()          # 하위 데이터 추출
df.index           # 데이터프레임 인덱스 보기
pd.RangeIndex(start=0, stop=891, step=1)
df.values          # 행, 열 구조 보기
df.sort_values(by=['범죄중분류'], axis=0)         # 구분별 기준 오름차순 정렬
```

```python
"""
3.	데이터 클린징

데이터 클린징 과정
이번 데이터의 유형은 범죄형 데이터이므로, 0의 값은 그 자체로 의미가 있다고 판단할 수 있습니다.
하지만, 각 유형의 범죄나 생활수준 구분에서, 미상, 기타 등 식별이 불가능한 데이터는 큰 의미가 없다고 판단,
또한 선거범죄, 환경범죄등 일상적 생활속 범죄와 관련이 없는 유형의 범죄역시 이번 조사에서 큰 의미가 없다고 판단하에,

다음과 같은 조건에 해당하는 데이터는 삭제하고 분석을 실시합니다.
    1) 기타, 미상등 각 유형, 카테고리에 해당되지 않은 미식별된 인덱스의 데이터
    2) 다음과 같이, 일상생활과 큰 연관이 없는 범죄유형을 삭제 - 증수뢰, 특별경제범죄,안보범죄,선거범죄,기타범죄
    3) 데이터프레임 분리를 위한 범죄대분류 삭제
"""

# null 값 제거(0도 없이 비어있는 값)
print(Police_crime_2015_backup.isnull().sum())
Police_crime_2015_cleaned = Police_crime_2015_backup.dropna()

# 필요 없는 열과 행 제거
Police_crime_2015_cleaned = Police_crime_2015_cleaned.drop(index=[19,28,34,35,37])


# 정제된 데이터 확인
print(Police_crime_2015_cleaned)
```

```python

"""
4.	데이터 탐색
"""

# 기본 통계 정보
print(Police_crime_2015_cleaned.describe())

# 데이터의 상관관계 확인
correlation_matrix = Police_crime_2015_cleaned.corr()
print(correlation_matrix)

# 상관관계 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('변수 간 상관관계 히트맵')
plt.show()

# 범죄 유형별 발생 빈도 확인
plt.figure(figsize=(14, 7))
sns.countplot(data=Police_crime_2015_cleaned, x='범죄중분류', order=Police_crime_2015_cleaned['범죄중분류'].value_counts().index)
plt.title('범죄 유형별 발생 빈도')
plt.xticks(rotation=90)
plt.show()

```

```python
"""
5.	데이터 분석
이번 데이터 분석의 핵심은 생활수준과 혼인관계로 인한 가정사와 발생되는 범죄의 연관성을 확인하는 것이 핵심입니다.
따라서, 각 생활수준과 혼인관계에 대한 데이터 분석을 실시한 뒤, 추가적으로 미혼자의 부모관계에 따른 범죄분석을 실시합니다.
"""
# 생활정도가 높은 모집군의 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='생활정도(상류)', y='범죄수', palette='Set9')
plt.title('생활정도가 높은 모집군의 범죄 발생 수')
plt.show()

# 생활정도가 중간인 모집군의 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='생활정도(중류)', y='범죄수', palette='Set8')
plt.title('생활정도가 중간인 모집군의 범죄 발생 수')
plt.show()

# 생활정도가 낮은 모집군의 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='생활정도(하류)', y='범죄수', palette='Set7')
plt.title('생활정도가 하류인 모집군의 범죄 발생 수')
plt.show()

# 혼인관계-배우자에 따른 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='혼인관계(유배우자)', y='범죄수', palette='Set6')
plt.title('혼인관계인 배우자가 있는 모집군의 범죄 발생 수')
plt.show()

# 혼인관계-동거에 따른 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='혼인관계(동거)', y='범죄수', palette='Set5')
plt.title('동거관계인 배우자가 있는 모집군의 범죄 발생 수')
plt.show()

# 혼인관계-이혼에 따른 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='혼인관계(이혼)', y='범죄수', palette='Set4')
plt.title('동거관계인 배우자가 있는 모집군의 범죄 발생 수')
plt.show()

# 혼인관계-사별에 따른 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='혼인관계(사별)', y='범죄수', palette='Set3')
plt.title('동거관계인 배우자가 있는 모집군의 범죄 발생 수')
plt.show()

# 미혼자의 양부부에 따른 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='미혼자의 부모관계(실(양)부모)', y='범죄수', palette='Set2')
plt.title('미혼자의 부모관계(실(양)부모)에 따른 범죄 발생 수')
plt.show()

# 미혼자의 미부모에 따른 범죄 발생 분석
plt.figure(figsize=(14, 7))
sns.boxplot(data=Police_crime_2015_cleaned, x='미혼자의 부모관계(무부모)', y='범죄수', palette='Set1')
plt.title('미혼자(무부모)에 따른 범죄 발생 수')
plt.show()
```

```python
"""
6.	데이터 시각화
"""
# 범죄 유형과 생활정도의 관계 시각화
plt.figure(figsize=(14, 7))
sns.barplot(data=Police_crime_2015_cleaned, x='범죄중분류', y='생활정도', ci=None)
plt.title('범죄 유형과 생활정도의 관계')
plt.xticks(rotation=90)
plt.show()

# 범죄 유형과 혼인관계의 관계 시각화
plt.figure(figsize=(14, 7))
sns.barplot(data=Police_crime_2015_cleaned, x='범죄중분류', y='혼인관계', ci=None)
plt.title('범죄 유형과 혼인관계의 관계')
plt.xticks(rotation=90)
plt.show()
```

```python
"""
7.	데이터 또는 분석 과정 차별화 포인트 
"""
# 범죄수에 따른 생활정도와 사회적 관계 비교
high_crime = Police_crime_2015_cleaned[Police_crime_2015_cleaned['범죄수'] > Police_crime_2015_cleaned['범죄수'].median()]
low_crime = Police_crime_2015_cleaned[Police_crime_2015_cleaned['범죄수'] <= Police_crime_2015_cleaned['범죄수'].median()]

plt.figure(figsize=(14, 7))
sns.boxplot(data=high_crime, x='생활정도', y='범죄수', palette='Set3')
plt.title('범죄수가 높은 그룹의 생활정도와 범죄수 관계')
plt.show()

plt.figure(figsize=(14, 7))
sns.boxplot(data=low_crime, x='생활정도', y='범죄수', palette='Set3')
plt.title('범죄수가 낮은 그룹의 생활정도와 범죄수 관계')
plt.show()
```
